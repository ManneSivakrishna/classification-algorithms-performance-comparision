# Define independent or dependent variable

```python
X = data2.drop(['target'],axis=1)
Y = data2['target']
```

# we are doing scaling by StandardScaler

```python
# Importing Sscaler by sklearn
from sklearn.preprocessing import StandardScaler 
# define our classifier
sc1 = StandardScaler()
# we are _fitted our classifier
ssc = sc1.fit_transform(X)
```

```python
# we are using TRAIN_TEST_SPLIT function for _split_details into TSTING & TAINING
X_tain, X_tst, Y_tain, Y_tst = train_test_split(ssc, Y, test_size=0.3, random_state=101)  #102
```

# using support vector classifier

```python
# we Import SVM Classifier using scikit learn library.
from sklearn.svm import SVC
# we are define SVC Classifier
svm1 = SVC(C=0.1)
# fitted our classifier
svm1.fit(X_tain, Y_tain)
```

```python
# we are doing prediction for classifier
prediction1 = svm1.predict(X_tst)
```

# Testing results for SVM

```python
# we are print here some results.

# we use metrics to print support vector classifier accuracy.accuracy_score
print('Accuracy SVM = ', metrics.accuracy_score(Y_tst, prediction1)*100) 
# using f1 score, we report the f1 score for support vector classifier.    
print('F1 score = ', f1_score(Y_tst, prediction1)*100) 
# precision score is used to output precision for the support vector classifier.                 
print('Precision = ', precision_score(Y_tst, prediction1)*100) 
# we use recall score to report the recall for support vector classifier.         
print('Recall = ', recall_score(Y_tst, prediction1)*100)   

acc1=metrics.accuracy_score(Y_tst, prediction1)*100
err1=100-acc1
print('Error rate =', err1)  # we print error rate for support vector classifier
```

# Training results for SVM

```python
# we are doing prediction for classifier
prediction11 = svm1.predict(X_tain)
```

```python
# we are print here some results.

# we use metrics to print support vector classifier accuracy.accuracy_score
print('Accuracy SVM = ', metrics.accuracy_score(Y_tain, prediction11)*100) 
# using f1 score, we report the f1 score for support vector classifier.    
print('F1 score = ', f1_score(Y_tain, prediction11)*100) 
# precision score is used to output precision for the support vector classifier.                 
print('Precision = ', precision_score(Y_tain, prediction11)*100) 
# we use recall score to report the recall for support vector classifier.         
print('Recall = ', recall_score(Y_tain, prediction11)*100)   
```

```python
# classification_report for svm classifier.
print(classification_report(Y_tst, prediction1))
```

```python
# Confugion_matix visual for SVM Classifier 
```

```python
svm1 = confusion_matrix(Y_tst, prediction1)

ac1 = sns.heatmap(svm1, annot=True, cmap='Oranges')

# fix the title for the matix
ac1.set_title('Confusion_matrix support vector classifier')
# fix X & Y lables for matix
ac1.set_xlabel('\n predictied labels')
ac1.set_ylabel('True labels')

PLOT.show()
```

# DecisionTreeClassifier (C4.5)
C4.5 basically a type of decision tree so we apply here decision tree in place of C4.5.

```python
# Import the Dtree classifier
from sklearn.tree import DecisionTreeClassifier 
# we define classifier
dre1 = DecisionTreeClassifier()
# fitted our Classifier
dre1.fit(X_tain, Y_tain)
```

# Testing results for Decision tree

```python
# we are print here some results.

# we use metrics.accuracy score to pint the accoracy of the d_tree Classifier.
print('Accuracy = ', metrics.accuracy_score(Y_tst, prediction2)*100)   
# for the d_tree classifer, we print the F1 Score.  
print('F1 score = ', f1_score(Y_tst, prediction2)*100)  
# precision score is used to output precision for a decision tree classifier.                
print('Precision = ', precision_score(Y_tst, prediction2)*100)   
# using reCall Score, we report the reCall for the decision tree classifier.       
print('Recall = ', recall_score(Y_tst, prediction2)*100) 

acc2=metrics.accuracy_score(Y_tst, prediction2)*100
err2=100-acc2
print('Error rate = ',err2)  # we print error rate for decision tree classifier
```
